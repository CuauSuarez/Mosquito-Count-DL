```
def getImages(listaCoor, api_key, folderPath, zoom_level = 15):
  for row in listaCoor:
    url = "https://maps.googleapis.com/maps/api/staticmap?center={}&zoom={}&size=400x400&maptype=satellite&key={}".format(row, zoom_level, api_key)
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    os.chdir(folderPath)
    img.save('{}.png'.format(row + "_zoom" + str(zoom_level)))
  return
```

```
folderPath = '/content/drive/My Drive/Mosquito-Tec/Images/Washington'
getImages(listaCoor, api_key, folderPath, zoom_level = 15)
getImages(listaCoor, api_key, folderPath, zoom_level = 16)
getImages(listaCoor, api_key, folderPath, zoom_level = 17)
```
